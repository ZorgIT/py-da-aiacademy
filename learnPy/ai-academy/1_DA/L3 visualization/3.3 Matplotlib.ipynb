````python
 import pandas as pd
 import matplotlib.pyplot as plt
 #matplotlib inline для корректного отображения на ноутбуке??
 df = pd.read_csv('Restaurant.csv')
 df.head()
````

````python
fig = plt.figure() #создать фигуру
axes = fig.add_axes([0, 0, 1, 1]) #добавляем в контейнер одну координатную плоскость, можно добавлять несколько
axes.hist(df["total_bill"], color = "green", bins = 20) # добавляем на координатную плоскость гистограмму с зп
#дооплнительно можно использовать опции color и bins - цвет и разбивка по диапазонам
 
# дополнительные параметры
"""
axes.set_title — заголовок диаграммы;
axes.set_ylabel — заголовок для оси ординат;
axes.set_xlabel — заголовок для оси абсцисс.
"""
axes.set_title("Общая сумма счета")
axes.set_ylabel("Количество заказов")
axes.set_xlabel("Сумма заказа, USD")
 
 
````

````python
#Диаграмма рассеяния
#Один признак по Х, а второй по Y
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.scatter(x = df["total_bill"],y = df["tip"], marker = "o") 
#Варианты маркеров https://matplotlib.org/stable/api/markers_api.html
axes.set_title('Общая сумма счёта Vs Cумма чаевых')
axes.set_xlabel('Сумма чека')
axes.set_ylabel('Размер чаевых')
````

````python
#Отображение данных из нескольких столбцов дф
%matplotlib inline

df = pd.read_csv('Restaurant.csv')

tips = df.pivot_table(values = ["total_bill", "tip"],
                      index = "day",
                      aggfunc = "mean")
display(tips.head())
````

````python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])

axes.bar (x = tips.index, height = tips ["total_bill"],
          width = 0.4, align = "edge")
axes.bar(x = tips.index, height=tips["tip"],
         width = -0.4, align = "edge",color ="red")

axes.set_title ("Средняя сумма счета Vs средняя сумма чаевых по дням недели")
axes.set_xlabel("Дни недели")
axes.set_ylabel("Сумма")
````

````python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])

axes.bar(x = tips.index, height = tips["total_bill"],
         width = 0.4, align = "edge", label = "Средняя сумма чека")
axes.bar(x = tips.index, height = tips['tip'],
         width = -0.4, align = "edge", label = "Средняя сумма чаевых")

axes.set_title("Средняя сумма счета Vs Средняя сумма чаевых по дням недели")
axes.set_xlabel("Дни недели")
axes.set_ylabel("Сумма")

axes.legend(loc = 1)
````

````python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.set_ylim(0,30) #Добавляем ограничение (вытягиваем ось y чтобы блок с легендой смотрелся нормально)

axes.bar(x = tips.index, height = tips["total_bill"],
         width = 0.4, align = "edge", label = "Средняя сумма чека")
axes.bar(x = tips.index, height = tips['tip'],
         width = -0.4, align = "edge", label = "Средняя сумма чаевых")

axes.set_title("Средняя сумма счета Vs Средняя сумма чаевых по дням недели")
axes.set_xlabel("Дни недели")
axes.set_ylabel("Сумма")

axes.legend(loc = 1)
````

### Метод Subplots

Можно  даже задать размер фигуры в дюймах (ширина*высота, 1 дюйм = 2,54 см) :

      fig, axes = plt.subplots(1, 3, figsize = (12,4))

После выполнения кода переменная fig содержит общий объект, а переменная axes содержит список, состоящий из трех внутренних объектов. Обращаться к элементам получившегося списка можно по индексам (от 0 до 2):
```
      fig, axes = plt.subplots(1, 3, figsize = (12,4))

      axes[0].hist(df.total_bill)
      axes[0].set_title('Общая сумма счета')

      axes[1].hist(df.tip)
      axes[1].set_title('Сумма чаевых')

      axes[2].scatter(x = df.total_bill, y = df.tip, marker = 'o')
      axes[2].set_title('Общая сумма счета Vs сумма чаевых')
```

````python
fig, axes = plt.subplots(1,3, figsize = (12,4))
````

````python

fig, axes = plt.subplots(1,3, figsize = (12,4))

axes[0].hist(df["total_bill"])
axes[0].set_title("Общая сумма счета")

axes[1].hist(df["tip"])
axes[1].set_title("Сумма чаевых")

axes[2].scatter(x = df["total_bill"], y = df["tip"], marker = "o")
axes[2].set_title("Общая сумма счета vs Средняя сумма чаевых")
````
