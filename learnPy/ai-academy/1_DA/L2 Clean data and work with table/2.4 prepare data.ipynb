### read_csv параметры
header = None — загрузить без строки с заголовком;

skiprows = n — пропустить n строк (часто у документов бывает техническая «шапка» или заголовки);

encoding — загрузить в конкретной кодировке;

na_values — список значений, который нужно заменить на NaN (специальный объект, обозначающий пропущенное значение).

```python
import pandas as pd
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

joined = pd.merge(movies, ratings, on='movieId', how='left')
joined.head()
```

```python
joined[joined.rating>4] #фильтрация по условию
```

```python
joined[(joined.movieId>4) & ("Drama" == joined.genres)] # комбинированные условия
```
