Ты — аналитик в рекламном отделе очень известной компании. Коллеги решили попробовать закупить рекламу в новом паблике и попросили тебя оценить активность его подписчиков. Ты решил сравнить количество лайков  в новом паблике со средним количеством лайков в пабликах, где вы закупаетесь обычно. Найди среднее арифметическое количества лайков в ваших обычных пабликах, а потом убедись, что оно корректно, вычислив разницу между медианой и ним. Округли результаты до десятых и выведи через пробел.



```python
import numpy as np
rating = np.array([[1,280,892,621],
                    [2,254,742,763],
                    [3,352,820,757],
                    [4,189,730,614],
                    [5,209,758,608],
                    [6,312,652,545],
                    [7,199,680,553],
                    [8,224,592,485],
                    [9,293,603,466],
                    [10,187,507,423]])
mean = np.mean(rating[:,2])
dif = np.median(rating[:,2])-np.mean(rating[:,2])
dif = round(dif, 1)
print(mean, dif)
```

Твои коллеги постоянно спорят о том, влияет ли количество постов на активность в сообществе, как связаны репосты и лайки и за чем “охотиться” в первую очередь. Ты решил доказать, что каждый показатель влияет на другие, и даже наименьший коэффициент корреляции будет довольно высок, поэтому важно отслеживать динамику каждого из показателей.

Определи коэффициент корреляции между двумя показателями в рейтинге пабликов из предыдущей задачи, связаных между собой меньше всего. Полученное значение корреляции округли до сотых и запиши результат в переменную c_min. Выведи результаты работы кода на экран.

```python
import numpy as np
rating = np.array([[1,280,892,621],
                    [2,254,742,763],
                    [3,352,820,757],
                    [4,189,730,614],
                    [5,209,758,608],
                    [6,312,652,545],
                    [7,199,680,553],
                    [8,224,592,485],
                    [9,293,603,466],
                    [10,187,507,423]])

cor_12=np.corrcoef(rating[:,1], rating[:,2])[0,1]
cor_13=np.corrcoef(rating[:,1], rating[:,3])[0,1]
cor_23=np.corrcoef(rating[:,2], rating[:,3])[0,1]

if cor_12<cor_13:
    if cor_12<cor_23:
        c_min = cor_12
    else:
        c_min = cor_23
else:
    if cor_13<cor_23:
        c_min = cor_13
    else:
        c_min = cor_23
c_min = round(c_min,2)
print(c_min)
```

Коллеги хотят запустить вирусный ролик и спорят, через какой из пабликов это лучше сделать: один говорит, что в пабликах нет особой разницы, поэтому можно выбирать по стоимости размещения, другой – что количество репостов значительно отличается от паблика к паблику и ориентироваться надо именно на этот показатель.

Выясни, насколько сильно отличаются паблики по количеству репостов. Для этого вычисли стандартное отклонение для общего количества репостов, полученное значение округли до сотых и запиши значение в переменную repost. Выведи результаты работы кода.

```python
import numpy as np
rating = np.array([[1,280,892,621],
                    [2,254,742,763],
                    [3,352,820,757],
                    [4,189,730,614],
                    [5,209,758,608],
                    [6,312,652,545],
                    [7,199,680,553],
                    [8,224,592,485],
                    [9,293,603,466],
                    [10,187,507,423]])
repost = np.std(rating[:,3])
repost = round(repost,2)
print(repost)
```
