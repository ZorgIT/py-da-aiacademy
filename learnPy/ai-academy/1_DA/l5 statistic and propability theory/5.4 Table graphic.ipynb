```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('Restaurant.csv')
df["sex"]=df["sex"].map({"Female":0, "Male":1}).astype(int)
df["vegan"]=df["vegan"].map({"No":0, "Yes":1}).astype(int)
df["time"]=df["time"].map({"Dinner":0, "Lunch":1}).astype(int)
df = df.drop(columns=["day"])

correlation = df.corr() #вычисляем коэффициенты корреляции между количественными признаками из датафрейма

sns.heatmap(correlation, annot = True, cmap ="coolwarm")
```
