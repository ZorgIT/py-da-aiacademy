Как устроен процесс анализа данных? Есть два основных подхода: статистический и визуальный. Визуальный анализ помогает увидеть общую картину исследования и оценить характер данных в датасете. С ним мы познакомились в 3 уроке. Теперь давай разбираться, что представляет собой статистический подход. 

Математическая статистика — это большой раздел математики, который профессиональные аналитики данных изучают годами. Нам с тобой будет достаточно основ, чтобы научиться делать верные выводы на основе информации из набора данных.

Статистический подход помогает проводить точные исследования, находить неочевидные связи внутри датасетов и ставить гипотезы.

Математически вероятность события равна отношению числа благоприятных событию исходов испытания к общему числу исходов испытания и обычно обозначается буквой P:

(P = frac{m}{N}) 
где m — это количество благоприятных исходов (например, количество победителей в лотерее), а N — общее количество исходов (общее количество людей, кто купил билеты, чтобы поучаствовать в лотерее). Представь конкурс в социальной сети с условием разыграть приз среди репостнувших конкурсную запись, тоже своего рода лотерея. Если разыгрывают 5 призов, а репост сделали 10 000 человек, то вероятность выиграть приз равна 0.0005. Устанешь репостить.

Теперь максимальное значение. Количество благоприятных исходов не может быть больше, чем общее количество исходов. Это как сказать, что в лотерее участвовало 50 человек, а выиграло 150, так не бывает. Самое большое число победителей — тоже 50 человек (если лотерея беспроигрышная). В такой ситуации получается, что количество благоприятных исходов совпадает с общим количеством исходов. Когда числитель совпадает со знаменателем в дроби, мы получаем единицу. Поэтому вероятность всегда находится в пределах от 0 до 1.

Закон распределения случайной величины — это соответствие между возможными значениями этой величины и их вероятностями. Чаще всего закон записывают таблицей, в которой перечислены в порядке возрастания все возможные значения случайной величины и соответствующие им вероятности:

(p + p_2 + p_3 + ... + p_n = 1 )


**Среднее арифметическое** — сумма значений, разделенная на их количество. Эта характеристика показывает одним числом обобщенное значение для данных. 

**Медиана (median)** — это число, которое делит упорядоченный ряд значений пополам. Значения в одной половине меньше медианы, а в другой — больше.

**взвешенное среднее арифметическое** — это среднее арифметическое для данных, где значения могут иметь разную «важность» или вес, и некоторые значения должны сильнее влиять на результат, чем другие.

# Показатели вариации
**Разброс данных** — это понятие, которое используют, чтобы описать отклонения значений набора данных от среднего. Начнем с математического ожидания.

**Математическое ожидание** — это среднее значение случайной величины.
Мы ожидаем увидеть это значение в будущем у новых значений, которые будут прибавляться к нашим данным. 


**Отклонение** — это разность между конкретным значением и средним арифметическим.

**Дисперсия** — сумма квадратов отклонений, деленная на их количество

Дисперсию саму по себе очень редко применяют на практике. Она удобна, чтобы вычислить стандартное отклонение для интересующего нас распределения значений. Делается это просто: берем дисперсию под корень и получаем нужное значение.

**Стандартное отклонение (std)** — корень из дисперсии. 
Чем меньше стандартное отклонение, тем значения конкретной характеристики однороднее. Чем больше стандартное отклонение, тем сильнее разброс значений признака.
