Прямоугольник — это и есть тот самый ящик (Box). Здесь собраны самые типичные значения, которые расположены между первым (Q1) и третьим (Q3) квартилем. Это делается для того, чтобы избежать искажений, отсекаются 25% самых маленьких значений и 25% самых больших значений, которые называются усами. То есть в нашем случае мы убираем 25% самых младших подписчиков и 25% самых взрослых подписчиков. Так мы сможем понять, какой в среднем возраст у типичного подписчика канала. По-другому «ящик» можно назвать межквартильным диапазоном (IQR - Interquartile Range).


```python
import pandas as pd
import seaborn as sns

restaraunt = pd.read_csv('Restaurant.csv')

sns.boxplot(x = "day",y="tip",data = restaraunt)
```

Ещё один способ увеличения информативности графика boxplot — использование параметра hue. Он дополнительно разделяет данные в зависимости от значения дополнительного категориального признака.


```python
sns.boxplot(x = "day",y="tip",data = restaraunt, hue="vegan")
```
