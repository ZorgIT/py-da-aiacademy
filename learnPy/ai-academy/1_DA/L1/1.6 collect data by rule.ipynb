```python
# Вычисляем среднюю зарплату
average_wage = football['Wage'].mean()

# Фильтруем футболистов с зарплатой выше средней
filtered_football = football[football['Wage'] > average_wage]

# Находим среднюю скорость для отфильтрованных данных
average_sprint_speed = filtered_football['SprintSpeed'].mean()
print(round(average_sprint_speed,2))
```

```python
max_composure = football['Composure'].max()
max_reactions = football['Reactions'].max()

# Фильтруем игроков и находим самого молодого
youngest_player_age = football.loc[
    (football['Composure'] > 0.9 * max_composure) & 
    (football['Reactions'] > 0.9 * max_reactions),
    'Age'
].min()

print(youngest_player_age)
```
