```python
import pandas as pd

football = pd.read_csv("football.csv")
display(football.head())
```

### Группировка объектов
Данные можно представить как последовательность единичных наблюдений, например, в нашем наборе данных единичное наблюдение — это футболист.

У каждого наблюдения есть атрибуты. Они могут быть разных типов:

категориальные — например, сборная, за которую играет футболист (признак Nationality);
численные — например, возраст футболиста (признак Age).
Задача группировки предполагает, что данные будут разделены по некоторому признаку (атрибуту), после чего к каждому элементу этих разделенных данных мы можем применить агрегирующую операцию. Это операция, которая позволит вычислить некоторый показатель.

Затем мы можем оценить, как отличаются эти показатели в зависимости от признака, по которому данные были разделены. Такое разделение мы назовем группировкой данных.

Возьмем по 5 футболистов трех различных национальностей, а затем посчитаем средний возраст внутри каждой из национальностей.



```python
groupA = football[football.Nationality == "Argentina"].Age
groupB = football[football.Nationality == "Brazil"].Age
groupC = football[football.Nationality == "Portugal"].Age
new_df = pd.DataFrame([groupA.mean(), groupB.mean(), groupC.mean()], columns = ["Age"], 
                      index = ["Argentina", "Brazil", "Portugal"])
display(new_df)
```
