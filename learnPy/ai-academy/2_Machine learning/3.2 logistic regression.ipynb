# Логистическая регрессия
Лог.рег. - метод обучения с учителем. определяет вероятность принадлежности к какой-то категории, используется для решения задач классификации.
Линейный классификатор — это гиперплоскость в пространстве признаков, которая разделяет объекты на подпространства по классам

По числу классов выделяют два вида классификации:

- Бинарная классификация — классификация, в которой есть всего 2 класса (например, класс котов и класс собак).
- Мультиклассовая классификация — классификация, в которой может быть 3 и более классов (например, определение болезни по симптомам).

## Описание логистической регрессии
Общая идея такая: есть функция f(х), где х — это точка данных из обучающей выборки. Если точка принадлежит классу «+», функция выдает некую вероятность. Если же точка принадлежит классу «-», то функция выдает значение, равное «1 - эта вероятность». То есть функция оценивает вероятность того, что модель правильно отнесла объект  выборки к определенному классу.  В процессе обучения логистической регрессии мы будем стараться максимизировать среднее значение этой функции, то есть увеличивать вероятность правильного определения класса.

Для точного предсказания мы будем использовать оценку по методу максимального правдоподобия (MLE — maximum likelihood estimation).

**Метод максимального правдоподобия** помогает выбрать лучшее предсказание через вычисление параметров распределения вероятностей.

пример формулы для отрицательных и 7 положительных ответах p∗(1−p)∗p∗(1−p)∗p∗p∗p∗p∗(1−p)∗p = p^7⋅(1−p)^3

Чтобы оценить качество классификатора, кроме метода максимальной правдоподобности используют отступ. **Отступ (margin)** объекта — буква \( \rho \) (читается как «ро») на рисунке ниже.

**Отступ** — степень «погруженности» объекта в свой класс.

Чем дальше отступ от нуля, тем больше уверенности в верном определении класса

Чем меньше значение отступа, тем ближе объект подходит к границе классов, и тем выше становится вероятность ошибки. Если отступ отрицательный, то модель ошибается и определяет класс неверно. Это ещё один вариант оценки качества классификатора.
