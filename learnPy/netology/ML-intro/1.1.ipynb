```python
import pandas as pd
import numpy as np
from skimage.metrics import mean_squared_error
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
df = pd.DataFrame(data, columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'])
df['MEDV'] = target
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
# Отделить колонки с признаками от колонки с целевой переменной (которую прогнозируем)
# это нужно для корректной работы sklearn
X = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y = df['MEDV']
```

```python
X
```

```python
y
```

```python
# Разделить данные на часть для обучения и часть для проверки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

### Делаем модель


```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

# Оценить качество модели на отложенных данных

```python
lr.predict(X_test)
```

```python
#оценка на глаз
pd.DataFrame((y_test.values, lr.predict(X_test))).T
```

```python
from sklearn.metrics import mean_squared_error
## Оцениваем при помощи MSE среднее квадратичное отклонение
mean_squared_error(y_test, lr.predict(X_test))
```

```python
#Оценка переобучения
mean_squared_error(y_train, lr.predict(X_train))
```
